```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
i=75
j=1
Read=str(i)+"_"+str(j)+".csv"
data=pd.read_csv(Read,skiprows=44).values
```

```python
Volts=data[3:1000,3]
Time=data[3:1000,1]
Volts = Volts.astype(np.float)
Time= Time.astype(np.float)
```

```python
v=(i*9.42477)/60
v=round(v,2)
label="Courbe de tension résiduelle à une vitesse de "+str(v)+ " cm/s"
m=2
plt.figure(figsize=(8,8))
for m in [2,3,4,5]:
    i=25*m
    j=1
    v=(i*9.42477)/60
    v=round(v,2)
    label="Vitesse de"+str(v)+" cm/s"
    Read=str(i)+"_"+str(j)+".csv"
    data=pd.read_csv(Read,skiprows=44).values
    Volts=data[3:1000,3]
    Time=data[3:1000,1]
    Volts = Volts.astype(np.float)
    Time= Time.astype(np.float)
    plt.subplot(4, 1, m-1)
    plt.plot(Time,Volts,linewidth=1)
    plt.xlabel('Temps (s)',fontsize=11)
    plt.ylabel('Tension (V)',fontsize=11)
    plt.title(str(v)+" cm/s")
plt.suptitle("Courbe de tension résiduelle pour 4 vitesses différentes")
plt.subplots_adjust(wspace = 0.5)
plt.subplots_adjust(hspace = 01.2)
plt.savefig('courbe_cc_vitesse_toute.eps')
```
